```python
data = []
import json
with open('multi_dataset.json') as f:
    data = json.load(f)
```

```python
def avg_num_answers(x):
    num = 0
    for triplet in x['multi_dataset']:
        num += len(triplet['answers'])
    return num/len(x['multi_dataset'])
```

```python
def len_question(x):
    length = 0
    for triplet in x['multi_dataset']:
        words = triplet['question'].split(' ')
        length += len(words)
    return length/len(x['multi_dataset'])
```

```python
def len_passage(x):
    length = 0
    for triplet in x['multi_dataset']:
        words = triplet['passage'].split(' ')
        length += len(words)
    return length/len(x['multi_dataset'])
```

```python
def len_answer(x):
#note: does not include 'noAnswer'
    length = 0
    answers = []
    for triplet in x['multi_dataset']:
        for answer in triplet['answers']:
            if answer != 'noAnswer':
                words = answer.split(' ')
                length += len(words)
                answers.append(answer)
    return length/len(answers)
```

```python
print('Average number of answers: '+str(avg_num_answers(data)))
print('Average question length: '+ str(len_question(data)))
print('Average passage length: '+ str(len_passage(data)))
print('Average answer length: '+ str(len_answer(data)))
print('Number of question/passage/answer triplets: '+str(len(data['multi_dataset'])))
```
