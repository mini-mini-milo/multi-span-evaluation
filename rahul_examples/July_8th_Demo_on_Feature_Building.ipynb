```python
import pandas as pd
```

```python
def convert_salary_to_float(salary):
    return float("".join(salary.split(",")))
```

```python
data = pd.read_csv("Employee_2.csv")
data["Annual_Salary"] = data["Annual_Salary"].apply(convert_salary_to_float)
```

```python
data.head()
```

```python
def gender_feature_function(row):
    if row["Gender"] == "Female":
        return 0
    return 1
```

```python
def salary_feature_function(row, mean, stdev):
    salary = row["Annual_Salary"]
    return (salary-mean)/stdev
```

```python
def still_employed(row):
    if str(row["Termination_Date"]) == "00/00/0000":
        return 0
    else:
        return 1
```

```python
mean_salary = data["Annual_Salary"].mean()
stdev_salary = data["Annual_Salary"].std()
```

```python
X = []
y = []
for index, row in data.iterrows():
    feature_vector = []
    feature_vector.append(gender_feature_function(row))
    feature_vector.append(salary_feature_function(row, mean_salary, stdev_salary))
    X.append(feature_vector)
    y.append(still_employed(row))
```
